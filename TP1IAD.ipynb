```python
#changement de dossier
import os
os.chdir("/content/data")
#importation des données
import pandas
dataframe = pandas.read_table("market_basket.txt",delimiter="\t",header=0)
#10 premières lignes
print(dataframe.head(10))
```

```python
#4 Table binaire
import numpy as np
col=dataframe['Product'].sort_values().unique()
lig=dataframe['ID'].unique()
matrix=np.zeros((len(lig),len(col)),int)
k=0
for i in lig:
  df=dataframe[dataframe["ID"]==i]
  a=0
  for j in col:
    if df['Product'].str.contains(j).any():
      matrix[k][a]=1
    a=a+1
  k=k+1

df = pandas.DataFrame(matrix, index=lig,columns=col)
df
```

```python
#7 la fonction a_priorie

import itertools
nbr_max_produit=2 #nbr max doit etre 4 mais ça sature la mémoire
def ExtractItemset(df, n):
    return list(itertools.combinations(df, n))

def CalculSupport (df,subsets,n):
  tab=[]
  k=0
  for i in subsets:
    subset=df[list(i)]
    w=subset[subset.sum(axis='columns')==n].count()
    tab.append(w[0])
    k=k+1
  return(tab)

#main
itemsetFrequent=[]
min_supp=0.025*len(lig)
for i in range(1,nbr_max_produit+1):
  itemset=ExtractItemset(col,i)
  Support=CalculSupport(df,itemset,i)
  ff= pandas.DataFrame(itemset)
  sup=pandas.DataFrame(Support)
  #print(ff[i-1])
  for j in range(0,len(itemset)):
    if(sup[0][j]>=min_supp):
      zz=[]
      for k in range(i):
        zz.append(ff[k][j])
      itemsetFrequent.append(zz)
  print(itemsetFrequent)
```

```python
#9 fonction is_inclus
def is_inclus(y,x):
  if y in x:
    return True;
  else:
    return False;
```

```python
#10 
for i in range(len(itemsetFrequent)):
  if is_inclus('Aspirin',itemsetFrequent[i]):
    print(itemsetFrequent[i]);
```

```python
#12

from mlxtend.frequent_patterns import apriori, association_rules

frq_items = apriori(df, min_support = 0.025, use_colnames = True)
rules = association_rules(frq_items, metric ="lift", min_threshold = 1)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
print(rules.head())
```
